```julia
using LinearAlgebra
using Plots
using CSV, DataFrames
using LaTeXStrings
```

```julia
function shift_rotate(x, y, θ, x_shift, y_shift, θ_shift)
    v = [x - x_shift, y-y_shift]
    R = [cos(θ_shift) sin(θ_shift); -sin(θ_shift) cos(θ_shift)]
    vf = R*v
    return vf[1], vf[2], (θ - θ_shift)
end
```

```julia
# This is the actual data-set
dubinsData = CSV.read("./DataFiles/twoCarsData.csv", DataFrame)

# Add mirrored data so that more data is available
dataMirrored = dubinsData

dataMirrored.yP = -dataMirrored.yP    # y = -y
dataMirrored.thetaP = -dataMirrored.thetaP    # th = -th
dataMirrored.inputP = -dataMirrored.inputP    # u = -u

dataMirrored.yE = -dataMirrored.yE    # y = -y
dataMirrored.thetaE = -dataMirrored.thetaE    # th = -th
dataMirrored.inputE = -dataMirrored.inputE    # u = -u

dubinsData = append!(dubinsData,dataMirrored)

dataP = dubinsData[:, [:xP, :yP, :thetaP, :xE, :yE, :thetaE, :inputP]]
dataE = dubinsData[:, [:xP, :yP, :thetaP, :xE, :yE, :thetaE, :inputE]]
dataP.inputP = Int.(round.(dataP.inputP))
dataE.inputE = Int.(round.(dataE.inputE))

# Transform dataP to pursuer's frame of reference
N = length(dataP.inputP)
for i=1:N
    dataP.xE[i], dataP.yE[i], dataP.thetaE[i] = shift_rotate(dataP.xE[i], dataP.yE[i], dataP.thetaE[i], dataP.xP[i], dataP.yP[i], dataP.thetaP[i])
    dataP.xP[i] = 0
    dataP.yP[i] = 0
    dataP.thetaP[i] = 0
end

# Transform dataE to evader's frame of reference
for i=1:N
    dataE.xP[i], dataE.yP[i], dataE.thetaP[i] = shift_rotate(dataE.xP[i], dataE.yP[i], dataE.thetaP[i], dataE.xE[i], dataE.yE[i], dataE.thetaE[i])
    dataE.xE[i] = 0
    dataE.yE[i] = 0
    dataE.thetaE[i] = 0
end

print(N)
```

## Pursuer Training

```julia
X_p = [dataP.xE dataP.yE dataP.thetaE]
y_p = dataP.inputP

# Split the data into test-train set
X_train, X_test, y_train, y_test = train_test_split(X_p, y_p, test_size = 0.20)

# naming the classifier Pursuer
Pursuer = RandomForestClassifier(n_estimators = 100)
Pursuer.fit(X_train, y_train)



dot_data = export_graphviz(Pursuer.estimators_[99], 
                           #feature_names=wine.feature_names,
                           #class_names=wine.target_names, 
                           filled=true, impurity=true, 
                           rounded=true)
println(dot_data)

y_pred = Pursuer.predict(X_test)

println("ACCURACY OF THE MODEL: ", accuracy_score(y_test, y_pred))
```

Simulation of pursuit-evasion game

```julia
# Simulation using the learned model
dubins_dynP(x, u) = [2.5*cos(x[3]), 2.5*sin(x[3]), 2.5*u]
dubins_dynE(x, u) = [1.0*cos(x[3]), 1.0*sin(x[3]), 1.0*u]


stateP0 = [0, 0, 0]
# stateP0 = [-5.5, -2.1, 2.78]

# stateE0 = [0, 7, -π/2]
stateE0 = [4.48, 3.92, -1.35] 
n_iter = 1500
dt = 0.01

statesP = zeros(n_iter, 3)
inputP = zeros(n_iter)
statesE = zeros(n_iter, 3)
inputE = zeros(n_iter)

statesP[1,:] .= stateP0
statesE[1,:] .= stateE0
last_ind = 1

for i=1:n_iter-1
    sP = statesP[i, :]
    sE = statesE[i, :]

    sET = shift_rotate(sE[1], sE[2], sE[3], sP[1], sP[2], sP[3])
    uP = Pursuer.predict([sET[1] sET[2] sET[3]])[1]
    statesP[i+1,:] .= statesP[i,:] .+ dt*dubins_dynP(statesP[i,:], uP[1])

    sPT = shift_rotate(sP[1], sP[2], sP[3], sE[1], sE[2], sE[3])
    uE = Evader.predict([sPT[1] sPT[2] sPT[3]])[1]
    statesE[i+1,:] .= statesE[i,:] .+ dt*dubins_dynE(statesE[i,:], uE[1])

    inputE[i] = uE

    last_ind = i
    if norm(sP[1:2] - sE[1:2]) < 0.1
        println("Captured!!")
        println(last_ind)
        println(uE)
        break
    end
end

```

Checking the Different sinusoidal Trajectories 

```julia
# Simulation using the learned model for pursuer, but evader follows sinusoidal trajectory
dubins_dynP(x, u) = [2*cos(x[3]), 2*sin(x[3]), 2*u]
dubins_dynE(x, u) = [1.0*cos(x[3]), 1.0*sin(x[3]), 1.0*u]

stateP0 = [0, 0, 0]
stateE0 = [3.0, -4.0, 0] 
n_iter = 1500
dt = 0.01

statesP = zeros(n_iter, 3)
inputP = zeros(n_iter)
statesE = zeros(n_iter, 3)
inputE = zeros(n_iter)

statesP[1,:] .= stateP0
statesE[1,:] .= stateE0
last_ind = 1

# Parameters for sinusoidal evader trajectory
sin_amplitude = 2.0    # Amplitude of sinusoidal motion
sin_frequency = 0.5    # Frequency of sinusoidal motion
sin_phase = 0.0        # Phase shift

for i=1:n_iter-1
    sP = statesP[i, :]
    sE = statesE[i, :]
    
    # Pursuer still uses its trained RF model
    sET = shift_rotate(sE[1], sE[2], sE[3], sP[1], sP[2], sP[3])
    uP = Pursuer.predict([sET[1] sET[2] sET[3]])[1]
    statesP[i+1,:] .= statesP[i,:] .+ dt*dubins_dynP(statesP[i,:], uP[1])
    
    # Evader follows sinusoidal trajectory instead of using RF model
    t = i * dt  # Current time
    
    # Method 1: Sinusoidal control input (oscillating turn rate)
    uE = sin_amplitude * sin(2π * sin_frequency * t + sin_phase)
    
    # Alternative Method 2: Predefined sinusoidal path (uncomment to use)
    # desired_heading = π/2 + sin_amplitude * sin(2π * sin_frequency * t + sin_phase)
    # heading_error = desired_heading - sE[3]
    # uE = 2.0 * atan(tan(heading_error/2))  # Bounded turn rate

    # Evader's sinusoidal motion depends on relative position to pursuer
    # t = i * dt
    relative_x = sP[1] - sE[1]
    relative_y = sP[2] - sE[2]
    distance_to_pursuer = norm([relative_x, relative_y])

    # Increase frequency when pursuer is close
    adaptive_frequency = sin_frequency * (1 + 2.0 * exp(-distance_to_pursuer))
    uE = sin_amplitude * sin(2π * adaptive_frequency * t)



    
    statesE[i+1,:] .= statesE[i,:] .+ dt*dubins_dynE(statesE[i,:], uE)
    
    inputE[i] = uE
    inputP[i] = uP[1]

    last_ind = i
    if norm(sP[1:2] - sE[1:2]) < 0.1
        println("Captured!!")
        println("Capture time: $(last_ind * dt) seconds")
        println("Capture at iteration: $last_ind")
        println("Final distance: $(norm(sP[1:2] - sE[1:2]))")
        break
    end
end

# Calculate final statistics
final_distance = norm(statesP[last_ind, 1:2] - statesE[last_ind, 1:2])
println("Final distance between pursuer and evader: $final_distance")
println("Simulation completed at iteration: $last_ind")
```

Checking the some intelligent Trajectories of evader

```julia
# Simulation using learned model for pursuer, with intelligent evader
dubins_dynP(x, u) = [2*cos(x[3]), 2*sin(x[3]), 2*u]
dubins_dynE(x, u) = [1.0*cos(x[3]), 1.0*sin(x[3]), 1.0*u]

stateP0 = [0, 0, 0]
stateE0 = [3.0, 3.0, π/2] 
n_iter = 2500
dt = 0.01

statesP = zeros(n_iter, 3)
inputP = zeros(n_iter)
statesE = zeros(n_iter, 3)
inputE = zeros(n_iter)

statesP[1,:] .= stateP0
statesE[1,:] .= stateE0
last_ind = 1

# Evader intelligence parameters
safety_distance = 3.0    # Distance at which evader starts evading aggressively
max_turn_rate = 1.0      # Maximum turning rate for evader
escape_gain = 2.0        # How aggressively the evader turns away

for i=1:n_iter-1
    sP = statesP[i, :]
    sE = statesE[i, :]
    
    # Pursuer uses trained RF model
    sET = shift_rotate(sE[1], sE[2], sE[3], sP[1], sP[2], sP[3])
    uP = Pursuer.predict([sET[1] sET[2] sET[3]])[1]
    statesP[i+1,:] .= statesP[i,:] .+ dt*dubins_dynP(statesP[i,:], uP[1])
    
    # INTELLIGENT EVADER: Calculate optimal escape maneuver
    distance_to_pursuer = norm(sP[1:2] - sE[1:2])
    
    if distance_to_pursuer < safety_distance
        # Calculate direction away from pursuer
        dx = sE[1] - sP[1]  # Vector from pursuer to evader
        dy = sE[2] - sP[2]
        
        # Desired escape direction (away from pursuer)
        desired_escape_angle = atan(dy, dx)
        
        # Current evader heading
        current_heading = sE[3]
        
        # Calculate angle difference (wrap to [-π, π])
        angle_diff = atan(sin(desired_escape_angle - current_heading), 
                         cos(desired_escape_angle - current_heading))
        
        # Control input: turn toward escape direction
        uE = escape_gain * angle_diff
        
        # Limit turning rate
        uE = clamp(uE, -max_turn_rate, max_turn_rate)
        
    else
        # If pursuer is far, move toward safe zone or random exploration
        # Move toward top-right corner (safe zone)
        target_x, target_y = 10.0, 10.0
        dx_target = target_x - sE[1]
        dy_target = target_y - sE[2]
        desired_angle = atan(dy_target, dx_target)
        
        angle_diff = atan(sin(desired_angle - sE[3]), cos(desired_angle - sE[3]))
        uE = 0.5 * angle_diff  # Less aggressive when not in immediate danger
    end
    
    statesE[i+1,:] .= statesE[i,:] .+ dt*dubins_dynE(statesE[i,:], uE)
    
    inputE[i] = uE
    inputP[i] = uP[1]

    last_ind = i
    if norm(sP[1:2] - sE[1:2]) < 0.1
        println("Captured!!")
        println("Capture time: $(last_ind * dt) seconds")
        println("Final distance: $(norm(sP[1:2] - sE[1:2]))")
        break
    end
end
```
